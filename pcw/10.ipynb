```python
import numpy as np
import matplotlib.pyplot as plt

def biasedCoin(n):
    path = [0]
    for _ in range(n):
        if np.random.rand() < 0.6:
            path.append(path[-1]+1)
        else:
            path.append(path[-1]-1)
    return path
```

```python
n = 30
plt.plot(range(n+1),biasedCoin(n))
plt.show()
```

```python
n = 30
trials = 200
samp = []
for _ in range(200):
    samp.append(biasedCoin(n)[-1])


print('mean: ', np.mean(samp))
print('var: ', np.var(samp))
plt.hist(samp)
plt.show()
```

```python
def gambRuinLimited(init, limit):
    R = [init]
    for _ in range(limit):
        w = np.random.randint(2)
        R.append(R[-1]+(2*w-1))
        if R[-1] < 0:
            break
    return R

for i in range(10000,81000,10000):
    samp = []
    for _ in range(100):
        samp.append(len(gambRuinLimited(100,i)))
    plt.hist(samp, label=(f'limit = {i}'))
    plt.legend()
    plt.show()
```
