```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

## 15.2

The number of egdes in a complete graph is $ \frac{n(n-1)}{2} $ because all n nodes have degree n-1, and every edge is counted twice

Similarly, a regular graph with degree k and n nodes has  $ \frac{nk}{2} $ edges


## 15.3

```python
complete_4 = nx.complete_graph(4)
complete_5 = nx.complete_graph(5)

biparte_23 = nx.Graph()
biparte_23.add_edges_from([(0,1), (0,2), (1,3), (2,3), (3,4)])

biparte_33 = nx.Graph()
biparte_33.add_edges_from([(0,1), (0,2), (0,5), (1,3), (1,4), (2,3), (2,4), (3,5), (4,5)])


nx.draw(complete_4)
plt.show()

nx.draw(complete_5)
plt.show()

nx.draw(biparte_23, node_color= ['green' if i in [0, 3] else 'blue' for i in biparte_23.nodes()])
plt.show()

nx.draw(biparte_33, node_color= ['green' if i in [0, 3, 4] else 'blue' for i in biparte_33.nodes()])
plt.show()
```
